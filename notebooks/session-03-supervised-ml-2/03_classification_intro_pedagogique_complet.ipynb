### üìä La Matrice de Confusion - Version Simple

Commen√ßons par un exemple ultra-simple avec 4 passagers :

**üéØ Exercice Interactif 2 :**

Pr√©disons manuellement la survie de ces 4 passagers :

1. **Marie** - 25 ans, 1√®re classe, femme ‚Üí Votre pr√©diction ?
2. **John** - 45 ans, 3√®me classe, homme ‚Üí Votre pr√©diction ?
3. **Emma** - 8 ans, 2√®me classe, fille ‚Üí Votre pr√©diction ?
4. **Paul** - 30 ans, 3√®me classe, homme ‚Üí Votre pr√©diction ?

*Notez vos pr√©dictions avant de continuer...*

````python
# Exercice interactif - Matrice de confusion simple
print("üéØ EXERCICE : Construisons une Matrice de Confusion")
print("=" * 60)

# Donn√©es de l'exercice
passagers = ["Marie (25, 1√®re, F)", "John (45, 3√®me, H)", "Emma (8, 2√®me, F)", "Paul (30, 3√®me, H)"]
vraies_survies = ["Survivant", "D√©c√©d√©", "Survivant", "D√©c√©d√©"]
predictions_typiques = ["Survivant", "Survivant", "Survivant", "D√©c√©d√©"]

print("\nüìã R√©sultats :")
print("Passager | Pr√©diction Typique | R√©alit√© | R√©sultat")
print("-" * 55)

tp = tn = fp = fn = 0

for i, (passager, pred, vrai) in enumerate(zip(passagers, predictions_typiques, vraies_survies)):
    if pred == "Survivant" and vrai == "Survivant":
        resultat = "TP ‚úÖ"
        tp += 1
    elif pred == "D√©c√©d√©" and vrai == "D√©c√©d√©":
        resultat = "TN ‚úÖ"
        tn += 1
    elif pred == "Survivant" and vrai == "D√©c√©d√©":
        resultat = "FP ‚ùå"
        fp += 1
    else:
        resultat = "FN ‚ùå"
        fn += 1
    
    print(f"{passager:20} | {pred:10} | {vrai:8} | {resultat}")

print(f"\nüìä Comptage :")
print(f"TP (True Positive) = {tp} - Correctement pr√©dit comme Survivant")
print(f"TN (True Negative) = {tn} - Correctement pr√©dit comme D√©c√©d√©") 
print(f"FP (False Positive) = {fp} - Incorrectement pr√©dit comme Survivant")
print(f"FN (False Negative) = {fn} - Incorrectement pr√©dit comme D√©c√©d√©")
````

### üìà Calculons les M√©triques Manuellement

Avec nos r√©sultats : TP=2, TN=1, FP=1, FN=0

**1. üéØ Accuracy (Pr√©cision Globale)**
```
Accuracy = (TP + TN) / (TP + TN + FP + FN)
Accuracy = (2 + 1) / (2 + 1 + 1 + 0) = 3/4 = 75%
```
**Interpr√©tation :** 75% de nos pr√©dictions sont correctes

**2. üîç Precision (Pr√©cision Positive)**
```
Precision = TP / (TP + FP)
Precision = 2 / (2 + 1) = 2/3 = 67%
```
**Interpr√©tation :** Parmi nos pr√©dictions "Survivant", 67% sont correctes

**3. üì° Recall (Rappel/Sensibilit√©)**
```
Recall = TP / (TP + FN)  
Recall = 2 / (2 + 0) = 2/2 = 100%
```
**Interpr√©tation :** Nous d√©tectons 100% des vrais survivants

**4. ‚öñÔ∏è F1-Score (√âquilibre)**
```
F1 = 2 √ó (Precision √ó Recall) / (Precision + Recall)
F1 = 2 √ó (0.67 √ó 1.00) / (0.67 + 1.00) = 80%
```
**Interpr√©tation :** Score √©quilibr√© entre Precision et Recall

````python
# Calcul automatique pour v√©rification
print("üîç V√âRIFICATION AUTOMATIQUE")
print("=" * 40)

accuracy = (tp + tn) / (tp + tn + fp + fn)
precision = tp / (tp + fp) if (tp + fp) > 0 else 0
recall = tp / (tp + fn) if (tp + fn) > 0 else 0
f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

print(f"Accuracy:  {accuracy:.1%}")
print(f"Precision: {precision:.1%}")
print(f"Recall:    {recall:.1%}")
print(f"F1-Score:  {f1:.1%}")
````

### üåä LA FONCTION SIGMO√èDE : LE C≈íUR DE LA CLASSIFICATION

#### ü§î Pourquoi avons-nous besoin de la sigmo√Øde ?

**Probl√®me :** Comment transformer n'importe quel nombre en probabilit√© ?
- Un score peut √™tre : -100, 0, 5.7, 1000...
- Une probabilit√© doit √™tre : entre 0 et 1 (0% √† 100%)

**Solution :** La fonction sigmo√Øde ! üåä

#### üìê La Formule Magique

    ```
    œÉ(x) = 1 / (1 + e^(-x))
    ```

**O√π :**
- `œÉ` (sigma) = fonction sigmo√Øde
- `x` = le score du mod√®le
- `e` = nombre d'Euler (‚âà 2.718)

#### üî¢ LE NOMBRE D'EULER : POURQUOI "e" ?

**Qu'est-ce que "e" ?**
- **Nom :** Nombre d'Euler (du math√©maticien Leonhard Euler)
- **Valeur :** e ‚âà 2.71828... (nombre irrationnel comme œÄ)
- **Symbole :** Souvent not√© "exp" dans les calculatrices

**ü§î Pourquoi utiliser "e" dans la sigmo√Øde ?**

**1. üìà Propri√©t√© de Croissance Naturelle**
- `e^x` est la fonction qui "grandit comme elle-m√™me"
- Sa d√©riv√©e est √©gale √† elle-m√™me : d/dx(e^x) = e^x
- Cela rend les calculs d'optimisation tr√®s √©l√©gants

**2. ‚ö° Calcul Efficace**
- Les ordinateurs calculent `e^x` tr√®s rapidement
- Algorithmes optimis√©s dans toutes les biblioth√®ques

**3. üéØ Forme Parfaite de la Courbe**
- `e^(-x)` d√©cro√Æt de fa√ßon "naturelle"
- Donne √† la sigmo√Øde sa forme en "S" parfaite
- Transition douce sans cassure

**üßÆ Visualisons l'Impact de "e" :**

**Comparaison avec d'autres bases :**
- Base 2 : `1/(1 + 2^(-x))` ‚Üí Transition trop abrupte
- Base 10 : `1/(1 + 10^(-x))` ‚Üí Transition trop brutale  
- Base e : `1/(1 + e^(-x))` ‚Üí Transition parfaite ! ‚ú®

**Interpr√©tation :**

- **Magnitude** ‚Üí Importance de la variable
